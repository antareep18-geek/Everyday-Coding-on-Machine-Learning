```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,accuracy_score
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/00488/Live_20210128.csv"
df=pd.read_csv(url)
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["Column1","Column2","Column3","Column4"], axis=1)
df.head()
```

```python
X=df.drop(columns=["status_id", "status_type","status_published"],axis=1)
y=pd.DataFrame(df["status_type"])
```

```python
y.value_counts()
```

```python
X.head()
```

```python
label_encoder=LabelEncoder()
y=label_encoder.fit_transform(y)
y.shape
```

```python
std_scaler=StandardScaler()
X=std_scaler.fit_transform(X)
```

```python
km=KMeans(n_clusters=2,random_state=10)
km.fit(X)
km.inertia_
```

```python
km=KMeans(n_clusters=4,random_state=10)
km.fit(X)
y_pred=km.predict(X)
cm=confusion_matrix(y,y_pred)
sum=0
for i in range(4):
  for j in range(4):
    if i==j:
      sum=sum+cm[i][j]
sum
```

```python
hac=AgglomerativeClustering(n_clusters=4,metric='euclidean', linkage='ward')
y_pred=hac.fit_predict(X)
y_pred[0]
```

```python
X.shape[0]
```

```python
score=accuracy_score(y,y_pred)*100
score
```
