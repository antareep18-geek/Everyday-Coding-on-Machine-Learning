```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import precision_score
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import log_loss
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("T124OPPE2_ModelBuilding_V1.csv")
df.shape
```

```python
df.columns
```

```python
X=df.iloc[:,:-1]
X.shape
```

```python
y=df.iloc[:,-1]
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, shuffle=False)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
perceptron=Perceptron(random_state=1729, max_iter=1, eta0=1, shuffle=False, fit_intercept=True, validation_fraction=0.1, warm_start=True, alpha=0 )
```

```python
for i in range(5):
  perceptron.fit(X_train,y_train)
  print("Bias ",i+1,"=",perceptron.intercept_)
```

```python
y_pred=perceptron.predict(X_train)
precision=precision_score(y_train,y_pred,pos_label=1)
precision
```

```python
logistic=SGDClassifier(loss='log_loss', penalty='l2', eta0=0.001, alpha=0, learning_rate='constant', random_state=1729, warm_start=True)
```

```python
for i in range(5):
  logistic.partial_fit(X_train,y_train, classes=[0,1])
  y_pred=logistic.predict_proba(X_train)
  loss=log_loss(y_train,y_pred)
  print("Loss in iteration",i+1,"=",loss)
```

```python
param_grid={
      'alpha':[0.0001, 0.0005, 0.001, 0.005],
      'eta0':[0.01, 0.05, 0.1, 0.5]
}
```

```python
grid=GridSearchCV(estimator=SGDClassifier(loss='log_loss', learning_rate='constant', random_state=1729),
                  param_grid=param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
new_logistic=SGDClassifier(alpha=0.0001,
                           eta0=0.01,
                           learning_rate='constant',
                           random_state=1729,
                           loss='log_loss',
                           class_weight={0:0.1,1:2}
                           )
```

```python
new_logistic.fit(X_train,y_train)
y_pred=new_logistic.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=new_logistic.classes_).plot()
```

```python
svm=SVC(kernel='rbf',
        decision_function_shape='ovr',
        random_state=1729,
        C=1)
```

```python
svm.fit(X_train,y_train)
y_pred=svm.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=svm.classes_).plot()
```

```python
tree=DecisionTreeClassifier(criterion='entropy',
                            splitter='random',
                            min_samples_split=4,
                            min_impurity_decrease=0.0001,
                            random_state=1729)
```

```python
tree.fit(X_train,y_train)
tree.get_depth(), tree.tree_.node_count
```

```python
plot_tree(tree, max_depth=1)
plt.show()
```

```python
bag1=BaggingClassifier(estimator=DecisionTreeClassifier(random_state=1729), random_state=1729, n_estimators=20)
bag2=BaggingClassifier(estimator=LogisticRegression(random_state=1729), random_state=1729, n_estimators=20)
bag3=BaggingClassifier(estimator=KNeighborsClassifier(), random_state=1729, n_estimators=20)
```

```python
bag1.fit(X_train,y_train)
train_score_1=bag1.score(X_train,y_train)
test_score_1=bag1.score(X_test,y_test)
print(abs(train_score_1-test_score_1))
```

```python
bag2.fit(X_train,y_train)
train_score_2=bag2.score(X_train,y_train)
test_score_2=bag2.score(X_test,y_test)
print(abs(train_score_2-test_score_2))
```

```python
bag3.fit(X_train,y_train)
train_score_3=bag3.score(X_train,y_train)
test_score_3=bag3.score(X_test,y_test)
print(abs(train_score_3-test_score_3))
```
