```python
# Define the conditions for dropping rows
invalid_livesin = ~df['LivesIn'].isin(['Village', 'City'])  # Invalid values in LivesIn
invalid_age = df['Age'] < 0                                # Negative values in Age
missing_bmi = df['BMI'].isnull()                           # Missing values in BMI

# Combine all conditions
invalid_rows = invalid_livesin | invalid_age | missing_bmi

# Drop rows with invalid values
df = df[~invalid_rows]
```
