```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
iris=load_iris(as_frame=True)
```

```python
X=iris.data
y=iris.target
```

```python
km=KMeans(init="random",
          n_init=50,
          max_iter=500,
          random_state=10,
          n_clusters=5)
```

```python
print(X['petal length (cm)'].corr(X['petal width (cm)']))
print(X['petal width (cm)'].corr(X['sepal length (cm)']))
print(X['petal length (cm)'].corr(X['sepal length (cm)']))
print(X['sepal width (cm)'].corr(X['sepal length (cm)']))
```

```python
km.fit(X,y)
km.inertia_
```

```python
k_range = range(1, 13)
inertias = []
for k in k_range:
    km = KMeans(n_clusters=k, init='random', n_init=50, max_iter=500, random_state=10)
    km.fit(X)  # Assuming X is your data
    inertias.append(km.inertia_)
plt.plot(k_range, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```
