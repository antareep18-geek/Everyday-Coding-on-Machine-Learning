```python
import pandas as pd
import numpy as np
from scipy.stats import poisson

# Load your data (replace with your file path if needed)
file_path = 'path/to/your/data_24t3.xlsx'
data = pd.read_excel("data_24t3.xlsx")

# Calculate mean of the observed data
mean_obs = data['obs'].mean()

# Observed frequencies for each unique count in "obs"
observed_counts = data['obs'].value_counts().sort_index()
observed_values = observed_counts.index
observed_frequencies = observed_counts.values

# Calculate expected frequencies based on Poisson distribution with mean of 3.92
expected_frequencies = [poisson.pmf(k, mean_obs) * len(data) for k in observed_values]

# For bins with expected frequencies < 5, combine them into a single bin
observed_frequencies_combined = []
expected_frequencies_combined = []
temp_obs = 0
temp_exp = 0

for obs_freq, exp_freq in zip(observed_frequencies, expected_frequencies):
    if exp_freq < 5:
        # Accumulate frequencies to combine low-frequency bins
        temp_obs += obs_freq
        temp_exp += exp_freq
    else:
        if temp_exp > 0:
            # Append the accumulated frequencies as a combined bin
            observed_frequencies_combined.append(temp_obs)
            expected_frequencies_combined.append(temp_exp)
            temp_obs = 0
            temp_exp = 0
        # Append current values if they are above the threshold
        observed_frequencies_combined.append(obs_freq)
        expected_frequencies_combined.append(exp_freq)

# Append any remaining accumulated bin
if temp_exp > 0:
    observed_frequencies_combined.append(temp_obs)
    expected_frequencies_combined.append(temp_exp)

# Calculate the chi-square test statistic
chi_square_stat = sum((obs - exp) ** 2 / exp for obs, exp in zip(observed_frequencies_combined, expected_frequencies_combined))

# Output the chi-square test statistic
chi_square_stat
```

```python
from scipy.stats import chi2

# Assume chi_square_stat is the computed chi-square statistic from previous code
# Calculate degrees of freedom (df)
df = len(observed_frequencies_combined) - 2  # one for mean, one for count constraint

# Calculate the p-value for the test
p_value = 1 - chi2.cdf(chi_square_stat, df)
p_value
```

```python
import pandas as pd

# Display the first few rows of the dataframe (optional)
print(data.head())

# Filter data for Female students
female_students = data[data['Gender'] == 'Female']

# Calculate total number of Female students
total_female = len(female_students)

# Calculate total number of students in Business subjects
business_students = data[data['Preference'] == 'Business']
total_business = len(business_students)

# Calculate expected value for Female students in Business subjects
expected_value_female_business = (total_female / len(data)) * total_business

# Print the expected value
print(f"Expected value of Female in Business subjects: {expected_value_female_business:.2f}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(data['Gender'], data['Preference'])

# Display the contingency table (optional)
print("Contingency Table:")
print(contingency_table)

# Calculate the Chi-square statistic and p-value
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
import pandas as pd
from scipy.stats import chi2


# Create a contingency table
contingency_table = pd.crosstab(data['Gender'], data['Preference'])

# Calculate degrees of freedom
num_rows = contingency_table.shape[0]  # Number of categories for Gender
num_columns = contingency_table.shape[1]  # Number of categories for Preference
dof = (num_rows - 1) * (num_columns - 1)

# Set the significance level
alpha = 0.05

# Get the tabulated Chi-square value
chi2_tabulated = chi2.ppf(1 - alpha, dof)

# Print the results
print(f"Tabulated Chi-square value at alpha = {alpha} and df = {dof}: {chi2_tabulated:.2f}")
```
