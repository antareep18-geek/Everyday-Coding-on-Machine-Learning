```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
import pandas as pd

# Assuming df is your DataFrame
X_test = X_test.drop(columns=['CATEGORY'])  # Step 1: Drop the CATEGORY column

# Define preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        # Step 2-3: CRIM and ZN: min-max scaling
        ('crim_minmax', MinMaxScaler(), ['CRIM']),
        ('zn_minmax', MinMaxScaler(), ['ZN']),

        # Step 4: INDUS: standard scaling
        ('indus_standard', StandardScaler(), ['INDUS']),

        # Step 5-11: POLINDEX, DIS, HIGHWAYCOUNT, TAX, PTRATIO, IMM, BPL: min-max scaling
        ('polindex_minmax', MinMaxScaler(), ['POLINDEX']),
        ('dis_minmax', MinMaxScaler(), ['DIS']),
        ('highwaycount_minmax', MinMaxScaler(), ['HIGHWAYCOUNT']),
        ('tax_minmax', MinMaxScaler(), ['TAX']),
        ('ptratio_minmax', MinMaxScaler(), ['PTRATIO']),
        ('imm_minmax', MinMaxScaler(), ['IMM']),
        ('bpl_minmax', MinMaxScaler(), ['BPL']),

        # Step 12: RM - impute with median then min-max scaling
        ('rm_pipeline', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler())
        ]), ['RM']),

        # Step 13: AGE - impute with mean then min-max scaling
        ('age_pipeline', Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', MinMaxScaler())
        ]), ['AGE']),

        # Step 14: RIVERSIDE - impute with most frequent value, then one-hot encode
        ('riverside_pipeline', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(drop='first'))
        ]), ['RIVERSIDE'])
    ],
    remainder='passthrough'  # Leave any other columns unchanged
)

# Apply the transformations to the DataFrame
X_test_transformed = preprocessor.fit_transform(X_test)

# Convert the result back to a DataFrame (optional, for easier inspection)
X_test_transformed = pd.DataFrame(X_test_transformed, columns=preprocessor.get_feature_names_out())

X_test_transformed.shape
```
