```python
import pandas as pd
import numpy as np
data = pd.read_csv("data_for_large_scale.csv")
```

```python
# Separating the features (X) and target (y)
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X.shape,y.shape
```

```python
X_array = np.array(X)
y_array = np.array(y)
X_array.shape,y_array.shape
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_array, y_array, test_size=0.3, random_state=10)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# Reshape X_train and X_test, but keep y_train and y_test as 1D arrays
import numpy as np

def reshape_features(data, samples=90):
    reshaped_data = data[:len(data) - len(data) % samples].reshape(-1, samples)
    return reshaped_data

X_train = reshape_features(X_train)
X_test = reshape_features(X_test)

# Truncate y_train and y_test to match reshaped X dimensions
y_train = y_train[:len(X_train)]
y_test = y_test[:len(X_test)]
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize SGD Regressor
sgd_regressor = SGDRegressor(random_state=10)

# Partial fit the model with training data
for i in range(1, 6):  # Perform 5 iterations
    sgd_regressor.partial_fit(X_train, y_train)

# Get coefficients and intercept
intercept = sgd_regressor.intercept_
coefficients = sgd_regressor.coef_

# Predictions
y_pred = sgd_regressor.predict(X_test)

# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Results
print("Intercept:", intercept)
print("Coefficients:", coefficients)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```
