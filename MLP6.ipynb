```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
X,y=fetch_openml('mnist_784', return_X_y=True)
X=X.to_numpy()
y=y.to_numpy()
```

```python
print(X.shape)
print(y.shape)
```

```python
X=MinMaxScaler().fit_transform(X)
```

```python
X_train=X[:10000]
y_train=y[:10000]
X_test=X[10000:12000]
y_test=y[10000:12000]
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train_5=[]
y_train_5=[]
X_train_3=[]
y_train_3=[]
for i in range (len(X_train)):
  if y_train[i]=='5':
    X_train_5.append(X_train[i])
    y_train_5.append(y_train[i])
  elif y_train[i]=='3':
    X_train_3.append(X_train[i])
    y_train_3.append(y_train[i])
```

```python
X_test_3_5=[]
y_test_3_5=[]
for i in range(len(X_test)):
  if y_test[i]=='3' or y_test[i]=='5':
    X_test_3_5.append(X_test[i])
    y_test_3_5.append(y_test[i])
X_test_3_5=np.array(X_test_3_5)
y_test_3_5=np.array(y_test_3_5)
y_test_3_5[y_test_3_5 == '5'] = 1
y_test_3_5[y_test_3_5 == '3'] = -1
print(X_test_3_5.shape)
print(y_test_3_5.shape)
```

```python
X_train_5 = np.array(X_train_5)
y_train_5 = np.array(y_train_5)
X_train_3 = np.array(X_train_3)
y_train_3 = np.array(y_train_3)

print(X_train_5.shape)
print(y_train_5.shape)
print(X_train_3.shape)
print(y_train_3.shape)

X_train_combined = np.vstack((X_train_5, X_train_3))
y_train_combined = np.hstack((y_train_5, y_train_3))
y_train_combined = y_train_combined.astype(int)

y_train_combined[y_train_combined == '5'] = 1
y_train_combined[y_train_combined == '3'] = -1

print(X_train_combined.shape)
print(y_train_combined.shape)

#Sanity check
print(f"Number of positive examples: {(y_train_combined == 1).sum()}")
print(f"Number of negative examples: {(y_train_combined == -1).sum()}")
```

```python
percep=Perceptron(random_state=42,eta0=1, max_iter=100, shuffle=True, fit_intercept=True,penalty=None)
percep.fit(X_train_combined,y_train_combined)
y_pred=percep.predict(X_test_3_5)
```

```python
cm=confusion_matrix(y_pred,y_test_3_5)
```

```python
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=percep.classes_).plot()
```

```python
# prompt: in this model calculate the accuracy, precision and recall score

from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test_3_5, y_pred)
precision = precision_score(y_test_3_5, y_pred, pos_label=1)
recall = recall_score(y_test_3_5, y_pred, pos_label=1)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
